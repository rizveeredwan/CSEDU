```python
import random
import csv
import pandas as pd 
import math
import numpy as np
```

```python
class LogisticRegression:
    def __init__(self,featureVectorNum,alpha):
            self.alpha=alpha
            self.featureVectorWeight=[]
            for i in range(0,featureVectorNum+1):
                self.featureVectorWeight.append(random.randint(1,10)/10.0) 
            self.featureVectorWeight = np.array(self.featureVectorWeight)
        
        
    def initiateProcess(self, trainFileName):
            xl = pd.ExcelFile(trainFileName)
            Sheet1 = xl.sheet_names
            #print Sheet1[0]
            self.trainDataframe=xl.parse(Sheet1[0])
            #print self.trainDataframe
    
            """
            #just for printing to access the dataframe
            for index,row in df1.iterrows():
                print(row['ID'],row['Reason for absence'],row['Month of absence'],row['Day of the week'],row['Seasons'])
            """

    def train(self):
            iteration=0
            while True:
                iteration=iteration+1
                if(iteration==1000):
                    break 
                cf=self.cost_function()
                print "cost function = ",cf
                self.theta_calculation()
            
    def cost_function(self):
            #function to calculate the cost function
            result=0.0
            feature_names=["","Temperature","Nausea","Lumbar pain","Urine pushing","Micturition pains","Burning of urethra","Nephritis of renal"]
            for index,row in self.trainDataframe.iterrows():
                vector=[]
                vector.append(1)
                for k in range(1,len(feature_names)):
                    vector.append(row[feature_names[k]])
                vector=np.array(vector)
                value= row['urinary bladder']*math.log10(self.htheta_function(vector))+(1-row['urinary bladder'])*math.log10(1-self.htheta_function(vector))
                result=result+value
            result=result*-1.0
            return result 

    def theta_calculation(self):
            updated_theta_values=[]
            feature_names=["","Temperature","Nausea","Lumbar pain","Urine pushing","Micturition pains","Burning of urethra","Nephritis of renal"]
            #theta-zero update 
            result=0
            for index,row in self.trainDataframe.iterrows():
                #making the feature vector
                vector=[]
                vector.append(1)
                for k in range(1,len(feature_names)):
                    vector.append(row[feature_names[k]])
                #calculating the value
                vector=np.array(vector)
                value = self.htheta_function(vector)
                #summing the result
                result=result+(value-row['urinary bladder'])
        
            #insert updated theta-zero value
            updated_theta_values.append(self.featureVectorWeight[0]-result*self.alpha*1.00)
            #update of other theta values 
            for i in range(1,len(self.featureVectorWeight)):
                result=0.0
                for index,row in self.trainDataframe.iterrows():
                    vector=[]
                    vector.append(1)
                    for k in range(1,len(feature_names)):
                        vector.append(row[feature_names[k]])
                    #calculating the value
                    vector=np.array(vector)
                    value= self.htheta_function(vector)
                    #summing the result
                    result=result+(value-row['urinary bladder'])*row[feature_names[i]]
                updated_theta_values.append(self.featureVectorWeight[i]-result*self.alpha*1.00)
            self.featureVectorWeight=np.array(updated_theta_values)
            return
    
    
    def htheta_function(self,vector):
            sumValue = np.dot(self.featureVectorWeight,vector)
            #applying e function 
            eValue = math.exp(-sumValue*1.0)
            #final result
            finalResult=1+eValue
            finalResult=1.0/(finalResult)
            if(finalResult == 1.0):
                finalResult=0.5
            return finalResult  
```
